```python
#importing required libraries

import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import re
import numpy as np
```

```python
url = "https://www.airlinequality.com/airline-reviews/american-airlines/page/"
```

```python
page = []
rev_rating = []
rev_head = []
rev_name = []
location = []
date_published = []
verified = []
review_txt = []
type_of_traveller = []
cabin_flown = []
route = []
date_flown = []
seat_comfort = []
cabin_staff_service = []
food_and_beverages = []
inflight_entertainment = []
ground_service = []
wifi_and_connectivity = []
value_for_money = []
recommended = []
```

```python
for p in range(1,344):
    urlx = url+'{}/'.format(p)
    # time.sleep(5)
    print(p)
    response = requests.get(urlx)
    content = BeautifulSoup(response.text, 'html')
#     print(content)
    Reviews = content.find("section",{"class":"layout-section layout-2 closer-top"})
#     print(Reviews)
    x = Reviews.find("article",{"class":"comp comp_reviews-airline querylist position-content"})
    for i in x.find_all("article",{"itemprop":"review"}):
        page.append(p)
        try:
            rev_rating.append(i.find("div",{"itemprop":"reviewRating"}).text)
        except:
            rev_rating.append(np.nan)
        rev_head.append(i.find("div",{"class":"body"}).find("h2",{"class":"text_header"}).text)
        rev_name.append(i.find("div",{"class":"body"}).find(itemprop="name").text)
        loc = i.find("div",{"class":"body"}).find(class_="text_sub_header userStatusWrapper").text
        if re.search(r'\((.*?)\)',loc):
            location.append(re.search(r'\((.*?)\)',loc).group(1))
        else:
            location.append(np.nan)
        date_published.append(i.find("div",{"class":"body"}).find(itemprop="datePublished").text)
        try:
            verified.append(i.find("div",{"class":"body"}).find(class_="tc_mobile").find(class_="text_content").find("strong").text)
        except AttributeError:
            verified.append(np.nan) # when it is not specified
        review_txt.append(i.find("div",{"class":"body"}).find(class_="tc_mobile").find(class_="text_content").text)
        stats = i.find("div",{"class":"body"}).find(class_="tc_mobile").find(class_="review-stats").find_all("tr")
        TypeOfTraveller = 0
        Cabin = 0
        Route=0
        Date_Flown=0
        Seat_Comfort = 0
        Cabin_Service=0
        Food_Beverages = 0
        Inflight_Ent=0
        Ground_serv=0
        Wifi=0
        Value_Money=0
        Recommend=0
        for k in stats:
            if k.find(class_="review-rating-header type_of_traveller"):
                TypeOfTraveller = k.find(class_="review-value").text
            elif k.find(class_="review-rating-header cabin_flown"):
                Cabin = k.find(class_="review-value").text
            elif k.find(class_="review-rating-header route"):
                Route = k.find(class_="review-value").text
            elif k.find(class_="review-rating-header date_flown"):
                Date_Flown = k.find(class_="review-value").text
            elif k.find(class_="review-rating-header seat_comfort"):
                try:
                    Seat_Comfort = max([int(h.text) for h in k.find(class_="review-rating-stars stars").find_all(class_="star fill")])
                except ValueError:
                    Seat_Comfort = 0
            elif k.find(class_="review-rating-header cabin_staff_service"):
                try:
                    Cabin_Service = max([int(h.text) for h in k.find(class_="review-rating-stars stars").find_all(class_="star fill")])
                except ValueError:
                    Cabin_Service = 0
            elif k.find(class_="review-rating-header food_and_beverages"):
                Food_Beverages = max([int(h.text) for h in k.find(class_="review-rating-stars stars").find_all(class_="star fill")])
            elif k.find(class_="review-rating-header inflight_entertainment"):
                try:
                    Inflight_Ent = max([int(h.text) for h in k.find(class_="review-rating-stars stars").find_all(class_="star fill")])
                except ValueError:
                    Inflight_Ent = 0
            elif k.find(class_="review-rating-header ground_service"):
                Ground_serv = max([int(h.text) for h in k.find(class_="review-rating-stars stars").find_all(class_="star fill")])
            elif k.find(class_="review-rating-header wifi_and_connectivity"):
                Wifi = max([int(h.text) for h in k.find(class_="review-rating-stars stars").find_all(class_="star fill")])
            elif k.find(class_="review-rating-header value_for_money"):
                Value_Money = max([int(h.text) for h in k.find(class_="review-rating-stars stars").find_all(class_="star fill")])
            elif k.find(class_="review-rating-header recommended"):
                Recommend = k.text
        if TypeOfTraveller:
            type_of_traveller.append(TypeOfTraveller)
        else:
            type_of_traveller.append(np.nan)
        if Cabin:
            cabin_flown.append(Cabin)
        else:
            cabin_flown.append(np.nan)
        if Route:
            route.append(Route)
        else:
            route.append(np.nan)
        if Date_Flown:
            date_flown.append(Date_Flown)
        else:
            date_flown.append(np.nan)
        if Seat_Comfort:
            seat_comfort.append(Seat_Comfort)
        else:
            seat_comfort.append(np.nan)
        if Cabin_Service:
            cabin_staff_service.append(Cabin_Service)
        else:
            cabin_staff_service.append(np.nan)
        if Food_Beverages:
            food_and_beverages.append(Food_Beverages)
        else:
            food_and_beverages.append(np.nan)
        if Inflight_Ent:
            inflight_entertainment.append(Inflight_Ent)
        else:
            inflight_entertainment.append(np.nan)
        if Ground_serv:
            ground_service.append(Ground_serv)
        else:
            ground_service.append(np.nan)
        if Wifi:
            wifi_and_connectivity.append(Wifi)
        else:
            wifi_and_connectivity.append(np.nan)
        if Value_Money:
            value_for_money.append(Value_Money)
        else:
            value_for_money.append(np.nan)
        if Recommend:
            recommended.append(Recommend)
        else:
            recommended.append(np.nan)
            
```

```python
Gayatri_airlines = pd.DataFrame({'Page':page, 'Rating':rev_rating, 'Review_Heading':rev_head, 
'Name':rev_name,'location':location, 'Date':date_published,'verified':verified,
'Review':review_txt, 'type_of_traveller':type_of_traveller,'cabin_flown':cabin_flown,'route':route,
'date_flown':date_flown,'seat_comfort':seat_comfort,'cabin_staff_service':cabin_staff_service,
'food_and_beverages':food_and_beverages,'inflight_entertainment':inflight_entertainment,'ground_service':ground_service,
'wifi_and_connectivity':wifi_and_connectivity,'value_for_money':value_for_money,'recommended':recommended})
```
