# Web Scraping- Amazon Best Sellers

#### Crawling through each Department in best sellers and scraping information

The same can be used for amazon "gift ideas", "new releases" tabs

```python
#importing required libraries

import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
url = "https://www.amazon.in/gp/bestsellers/"
```

```python
response = requests.get(url)
response
```

```python
content = BeautifulSoup(response.text, 'html')
```

```python
type(content)
```

```python
best_seller = content.find(id="zg_left_col2")
```

```python
best_seller_dict= {}
for i in best_seller.find_all("a"):
    best_seller_dict[i.text] = i["href"]
best_seller_dict
```
